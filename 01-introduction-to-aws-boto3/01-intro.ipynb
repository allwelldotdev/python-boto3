import `boto3` and create a configuration session to specify aws access info


```python
import boto3

session = boto3.Session(
    profile_name="boto3_101"
)  # accessing aws via boto3 using a specific profile (aws user) in `~/.aws/credentials` file
```

```python
s3 = session.resource("s3")  # grabbing the s3 resource
```

```python
list(s3.buckets.all())  # iterating through s3 buckets
```

```python
for bucket in s3.buckets.all():
    print(bucket.name)  # using for loop to iterate and print bucket name attribute
```
